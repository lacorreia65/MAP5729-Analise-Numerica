# Parte 1 - Método da bissecção

Seja $f: \mathbb{R} \to \mathbb{R}$ e $a,b \in \mathbb{R}$ tais que $f(a) f(b) < 0$. Se $f$ é contínua em $[a,b]$, sabemos, pelo Teorema do Valor Intermediário, que existe uma raiz $\xi$ de $f$ em $(a,b)$. O método da bissecção consiste em determinar iterativamente intervalos $[a_i,b_i]$, cada vez menores, nos quais $\xi$ está contido. Isso é feito da seguinte forma: começando com o intervalo $[a,b]$:


1. Calcule $f$ no ponto médio do intervalo, $x_m = \frac{a+b}{2} $  
2. Se $f(x_m) = 0$ ou algum critério de parada é satisfeito, termine o processo iterativo, sendo $x_m$ a aproximação obtida para $\xi$
3. Se $f(x_m) f(x_a) < 0$, então $\xi$ está em $(a,x_m)$, e definimos o novo intervalo $(a,x_m)$ (ou seja, $b$ recebe $x_m$). Caso contrário, $\xi$ está em $(x_m,b)$, e definimos o novo intervalo $(x_m,b)$ (ou seja, $a$ recebe $x_m$).
